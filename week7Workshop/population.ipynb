```python
import matplotlib.pyplot as plt

# Parameters
k = 0.1  # Reproduction rate
K = 1000  # Carrying capacity
m = 10  # Catch rate
total_time = 52 * 5  # 5 years in weeks
delta_t = 1  # Time step in weeks

# Initial populations to simulate
initial_populations = [300, 600, 900, 1200, 1500]

# Function to compute the derivative dP/dt
def derivative_PT(k, P, K, m):
    return k * P * (1 - P / K) - m

# Simulation for each initial population
plt.figure(figsize=(10, 6))
for P0 in initial_populations:
    iterative_solutions = [P0]  # Start with the initial population
    p_n_prev = P0  # Initialize the previous population value
    time_steps = [0]  # Track time steps for plotting

    # Run the simulation
    for t in range(1, total_time + 1):
        p_n = p_n_prev + delta_t * derivative_PT(k, p_n_prev, K, m)
        if p_n <= 0:  # Stop if population drops to zero or below
            break
        iterative_solutions.append(p_n)
        p_n_prev = p_n
        time_steps.append(t)

    # Plot the results for this initial population
    plt.plot(time_steps, iterative_solutions, label=f'P(0) = {P0}')

# Add labels, title, and legend
plt.title('Fish Population Over Time')
plt.xlabel('Time (days)')
plt.ylabel('Population P(t)')
plt.grid(True)
plt.legend()
plt.show()
```
